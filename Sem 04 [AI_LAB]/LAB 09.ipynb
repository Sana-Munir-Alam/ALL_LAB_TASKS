```python
print("\n================ TASK 1 ================")

total_cards = 52
red_cards = 26
hearts = 13
diamonds = 13

print("\n(1) Probability of red card:")
p_red = red_cards / total_cards
print("P(Red) =", p_red)

print("\n(2) Given red card, probability of heart:")
p_heart_given_red = hearts / red_cards
print("P(Heart | Red) =", p_heart_given_red)

print("\n(3) Given face card, probability of diamond:")
face_cards = 12
diamond_face = 3
p_diamond_given_face = diamond_face / face_cards
print("P(Diamond | Face) =", p_diamond_given_face)

print("\n(4) Given face card, probability that it is a spade or a queen:")
spade_face = 3
queen_face = 4
queen_of_spades = 1
p_spade_or_queen_given_face = (spade_face + queen_face - queen_of_spades) / face_cards
print("P(Spade or Queen | Face) =", p_spade_or_queen_given_face)
```

```python
print("\n================ TASK 2 ================")

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

model = DiscreteBayesianNetwork([
    ('Intelligence', 'Grade'),
    ('StudyHours', 'Grade'),
    ('Difficulty', 'Grade'),
    ('Grade', 'Pass')
])

cpd_i = TabularCPD('Intelligence', 2, [[0.7], [0.3]])
cpd_s = TabularCPD('StudyHours', 2, [[0.6], [0.4]])
cpd_d = TabularCPD('Difficulty', 2, [[0.4], [0.6]])

cpd_g = TabularCPD(
    'Grade', 3,
    values=[
        [0.90, 0.70, 0.60, 0.40, 0.30, 0.20, 0.10, 0.05],
        [0.08, 0.20, 0.30, 0.35, 0.40, 0.50, 0.50, 0.35],
        [0.02, 0.10, 0.10, 0.25, 0.30, 0.30, 0.40, 0.60]
    ],
    evidence=['Intelligence', 'StudyHours', 'Difficulty'],
    evidence_card=[2, 2, 2]
)

cpd_p = TabularCPD(
    'Pass', 2,
    values=[
        [0.05, 0.20, 0.50],
        [0.95, 0.80, 0.50]
    ],
    evidence=['Grade'],
    evidence_card=[3]
)

model.add_cpds(cpd_i, cpd_s, cpd_d, cpd_g, cpd_p)
assert model.check_model()

infer = VariableElimination(model)

print("\n(1) P(Pass | StudyHours=Sufficient, Difficulty=Hard)")
res1 = infer.query(variables=['Pass'], evidence={'StudyHours': 0, 'Difficulty': 0})
print(res1)

print("\n(2) P(Intelligence | Pass=Yes)")
res2 = infer.query(variables=['Intelligence'], evidence={'Pass': 1})
print(res2)
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

print("\n================ TASK 3 ================")

# STRUCTURE
model = DiscreteBayesianNetwork([
    ('Disease', 'Fever'),
    ('Disease', 'Cough'),
    ('Disease', 'Fatigue'),
    ('Disease', 'Chills')
])

# PRIOR
cpd_d = TabularCPD('Disease', 2, [[0.3], [0.7]])  # Flu, Cold

# CPTs
cpd_fever = TabularCPD(
    'Fever', 2,
    values=[[0.1, 0.5],
            [0.9, 0.5]],
    evidence=['Disease'],
    evidence_card=[2]
)

cpd_cough = TabularCPD(
    'Cough', 2,
    values=[[0.2, 0.4],
            [0.8, 0.6]],
    evidence=['Disease'],
    evidence_card=[2]
)

cpd_fatigue = TabularCPD(
    'Fatigue', 2,
    values=[[0.3, 0.7],
            [0.7, 0.3]],
    evidence=['Disease'],
    evidence_card=[2]
)

cpd_chills = TabularCPD(
    'Chills', 2,
    values=[[0.4, 0.6],
            [0.6, 0.4]],
    evidence=['Disease'],
    evidence_card=[2]
)

model.add_cpds(cpd_d, cpd_fever, cpd_cough, cpd_fatigue, cpd_chills)

assert model.check_model()

infer = VariableElimination(model)

# TASK 1
print("\n(1) P(Disease | Fever=Yes, Cough=Yes)")
res1 = infer.query(
    variables=['Disease'],
    evidence={'Fever': 1, 'Cough': 1}
)
print(res1)

# TASK 2
print("\n(2) P(Disease | Fever=Yes, Cough=Yes, Chills=Yes)")
res2 = infer.query(
    variables=['Disease'],
    evidence={'Fever': 1, 'Cough': 1, 'Chills': 1}
)
print(res2)

# TASK 3
print("\n(3) P(Fatigue | Disease=Flu)")
res3 = infer.query(
    variables=['Fatigue'],
    evidence={'Disease': 0}
)
print(res3)
```

```python
import numpy as np

print("\n================ TASK 4 ================")

states = ["Sunny", "Cloudy", "Rainy"]

transition_matrix = np.array([
    [0.6, 0.3, 0.1],
    [0.3, 0.4, 0.3],
    [0.2, 0.3, 0.5]
])

def simulate(days, start_state="Sunny"):
    current = start_state
    rainy_count = 0
    
    for _ in range(days):
        if current == "Sunny":
            probs = transition_matrix[0]
        elif current == "Cloudy":
            probs = transition_matrix[1]
        else:
            probs = transition_matrix[2]
        
        next_state = np.random.choice(states, p=probs)
        
        if next_state == "Rainy":
            rainy_count += 1
        
        current = next_state
    
    return rainy_count

days = 10
num_simulations = 100000
np.random.seed(42)

rainy_counts = []
for _ in range(num_simulations):
    rain_count = simulate(days)
    rainy_counts.append(rain_count)

rainy_counts = np.array(rainy_counts)
prob_at_least_3_rainy = np.sum(rainy_counts >= 3) / num_simulations

print(f"\nProbability of at least 3 rainy days in {days} days: {prob_at_least_3_rainy:.4f}")

sequence_example = []
current = "Sunny"
sequence_example.append(current)
for _ in range(days):
    if current == "Sunny":
        probs = transition_matrix[0]
    elif current == "Cloudy":
        probs = transition_matrix[1]
    else:
        probs = transition_matrix[2]
    next_state = np.random.choice(states, p=probs)
    sequence_example.append(next_state)
    current = next_state

print("\nExample weather sequence (first 10 days plus start):")
print(" -> ".join(sequence_example))
```
