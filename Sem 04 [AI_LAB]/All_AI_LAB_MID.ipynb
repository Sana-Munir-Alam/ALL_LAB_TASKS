````python
# Breadth First Search

from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def is_goal(self, node):
        return node == self.goal


class BFSAgent:
    def action(self, environment):
        queue = deque()                  # FIFO Queue
        queue.append([environment.start])  # Each element is a PATH
        visited = set()

        while queue:
            path = queue.popleft()       # Take from FRONT
            node = path[-1]              # Current node = last in path

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | Path so far: {path}")

            if environment.is_goal(node):
                return path              # Return full path to goal

            for neighbor in environment.get_neighbors(node):
                if neighbor not in visited:
                    queue.append(path + [neighbor])

        return None                      # No path found

    def run_agent(self, environment):
        print("=== Breadth First Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = BFSAgent()
agent.run_agent(env)
````

````python
# Depth First Search

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def is_goal(self, node):
        return node == self.goal


class DFSAgent:
    def action(self, environment):
        stack = [[environment.start]]    # Stack of PATHS
        visited = set()

        while stack:
            path = stack.pop()           # Take from TOP (LIFO)
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | Path so far: {path}")

            if environment.is_goal(node):
                return path

            for neighbor in environment.get_neighbors(node):
                if neighbor not in visited:
                    stack.append(path + [neighbor])

        return None

    def run_agent(self, environment):
        print("=== Depth First Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = DFSAgent()
agent.run_agent(env)
````

````python
# Depth Limited Search

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def is_goal(self, node):
        return node == self.goal


class DLSAgent:
    def action(self, environment, limit):
        # Stack stores (path, depth)
        stack = [([environment.start], 0)]
        visited = set()

        while stack:
            path, depth = stack.pop()
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | Depth={depth} | Path: {path}")

            if environment.is_goal(node):
                return path

            if depth < limit:            # Only go deeper if within limit
                for neighbor in environment.get_neighbors(node):
                    if neighbor not in visited:
                        stack.append((path + [neighbor], depth + 1))
            else:
                print(f"  ✗ Depth limit {limit} reached at {node}, pruning.")

        return None

    def run_agent(self, environment, limit=2):
        print(f"=== Depth Limited Search (Limit={limit}) ===")
        result = self.action(environment, limit)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print(f"\n✗ Goal not found within depth {limit}.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = DLSAgent()
agent.run_agent(env, limit=2)
````

````python
# IDDFS

graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def is_goal(self, node):
        return node == self.goal


class IDDFSAgent:
    def dls(self, environment, limit):
        stack = [([environment.start], 0)]
        visited = set()

        while stack:
            path, depth = stack.pop()
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"    Visiting: {node} | Depth={depth}")

            if environment.is_goal(node):
                return path

            if depth < limit:
                for neighbor in environment.get_neighbors(node):
                    if neighbor not in visited:
                        stack.append((path + [neighbor], depth + 1))
        return None

    def action(self, environment, max_depth=5):
        for limit in range(max_depth + 1):      # Increase limit each iteration
            print(f"\n  >> Trying depth limit = {limit}")
            result = self.dls(environment, limit)
            if result:
                return result
        return None

    def run_agent(self, environment):
        print("=== Iterative Deepening DFS ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = IDDFSAgent()
agent.run_agent(env)
````

````python
# UCS

import heapq

# Weighted graph: (neighbor, cost)
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def is_goal(self, node):
        return node == self.goal


class UCSAgent:
    def action(self, environment):
        # Priority queue: (cost, path)
        pq = [(0, [environment.start])]
        visited = set()

        while pq:
            cost, path = heapq.heappop(pq)   # Pop LOWEST cost
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | Cost={cost} | Path: {path}")

            if environment.is_goal(node):
                return path, cost

            for neighbor, edge_cost in environment.get_neighbors(node):
                if neighbor not in visited:
                    new_cost = cost + edge_cost
                    heapq.heappush(pq, (new_cost, path + [neighbor]))

        return None, float('inf')

    def run_agent(self, environment):
        print("=== Uniform Cost Search ===")
        result, cost = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result} | Total Cost: {cost}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = UCSAgent()
agent.run_agent(env)
````

````python
# Bi-directional

from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B'],
    'F': ['C']
}

class Environment:
    def __init__(self, graph, start, goal):
        self.graph = graph
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])


class BidirectionalAgent:
    def action(self, environment):
        # Forward search from start
        front_queue = deque([[environment.start]])
        front_visited = {environment.start: [environment.start]}

        # Backward search from goal
        back_queue = deque([[environment.goal]])
        back_visited = {environment.goal: [environment.goal]}

        while front_queue or back_queue:

            # --- Forward Step ---
            if front_queue:
                path = front_queue.popleft()
                node = path[-1]
                print(f"  Forward visiting: {node}")

                for neighbor in environment.get_neighbors(node):
                    if neighbor not in front_visited:
                        new_path = path + [neighbor]
                        front_visited[neighbor] = new_path
                        front_queue.append(new_path)

                        # Check if forward met backward
                        if neighbor in back_visited:
                            full_path = new_path + back_visited[neighbor][-2::-1]
                            return full_path

            # --- Backward Step ---
            if back_queue:
                path = back_queue.popleft()
                node = path[-1]
                print(f"  Backward visiting: {node}")

                for neighbor in environment.get_neighbors(node):
                    if neighbor not in back_visited:
                        new_path = path + [neighbor]
                        back_visited[neighbor] = new_path
                        back_queue.append(new_path)

                        # Check if backward met forward
                        if neighbor in front_visited:
                            full_path = front_visited[neighbor] + new_path[-2::-1]
                            return full_path

        return None

    def run_agent(self, environment):
        print("=== Bidirectional Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Meeting point found! Path: {result}")
        else:
            print("✗ No path found.")


# --- Run ---
env = Environment(graph, 'A', 'F')
agent = BidirectionalAgent()
agent.run_agent(env)
````

````python
# Best First Search

import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 7,
    'E': 5,
    'F': 0
}

class Environment:
    def __init__(self, graph, heuristic, start, goal):
        self.graph = graph
        self.heuristic = heuristic
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def get_heuristic(self, node):
        return self.heuristic.get(node, float('inf'))

    def is_goal(self, node):
        return node == self.goal


class BestFirstSearchAgent:
    def action(self, environment):
        # Priority queue ordered by h(n) only
        # (heuristic, path)
        pq = [(environment.get_heuristic(environment.start),
               [environment.start])]
        visited = set()

        while pq:
            h, path = heapq.heappop(pq)    # Pop node with LOWEST h(n)
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | h(n)={h} | Path: {path}")

            if environment.is_goal(node):
                return path

            for neighbor, _ in environment.get_neighbors(node):
                if neighbor not in visited:
                    h_val = environment.get_heuristic(neighbor)
                    heapq.heappush(pq, (h_val, path + [neighbor]))

        return None

    def run_agent(self, environment):
        print("=== Best First Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, heuristic, 'A', 'F')
agent = BestFirstSearchAgent()
agent.run_agent(env)
````

````python
# Greedy Best First Search

import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 7,
    'E': 5,
    'F': 0
}

class Environment:
    def __init__(self, graph, heuristic, start, goal):
        self.graph = graph
        self.heuristic = heuristic
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def get_heuristic(self, node):
        return self.heuristic.get(node, float('inf'))

    def is_goal(self, node):
        return node == self.goal


class GreedyBestFirstAgent:
    def action(self, environment):
        # (h(n), path) — greedy: only h(n), no actual cost
        pq = [(environment.get_heuristic(environment.start),
               [environment.start])]
        visited = set()

        while pq:
            h, path = heapq.heappop(pq)
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            print(f"  Visiting: {node} | h(n)={h} | Path: {path}")

            if environment.is_goal(node):
                return path

            neighbors = environment.get_neighbors(node)

            # Sort neighbors by heuristic — greedily pick lowest h
            neighbors_sorted = sorted(
                neighbors,
                key=lambda x: environment.get_heuristic(x[0])
            )
            print(f"  Neighbors sorted by h: "
                  f"{[(n, environment.get_heuristic(n)) for n,_ in neighbors_sorted]}")

            for neighbor, _ in neighbors_sorted:
                if neighbor not in visited:
                    h_val = environment.get_heuristic(neighbor)
                    heapq.heappush(pq, (h_val, path + [neighbor]))

        return None

    def run_agent(self, environment):
        print("=== Greedy Best First Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, heuristic, 'A', 'F')
agent = GreedyBestFirstAgent()
agent.run_agent(env)
````

````python
# A* Search

import heapq

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 1)],
    'D': [],
    'E': [],
    'F': []
}

heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 7,
    'E': 5,
    'F': 0
}

class Environment:
    def __init__(self, graph, heuristic, start, goal):
        self.graph = graph
        self.heuristic = heuristic
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def get_heuristic(self, node):
        return self.heuristic.get(node, float('inf'))

    def is_goal(self, node):
        return node == self.goal


class AStarAgent:
    def action(self, environment):
        start = environment.start
        h_start = environment.get_heuristic(start)

        # (f(n), g(n), path)
        pq = [(h_start, 0, [start])]
        visited = set()

        while pq:
            f, g, path = heapq.heappop(pq)   # Pop lowest f(n)
            node = path[-1]

            if node in visited:
                continue
            visited.add(node)

            h = environment.get_heuristic(node)
            print(f"  Visiting: {node} | g={g}, h={h}, f={f} | Path: {path}")

            if environment.is_goal(node):
                return path, g               # g = total actual cost

            for neighbor, edge_cost in environment.get_neighbors(node):
                if neighbor not in visited:
                    new_g = g + edge_cost                        # Actual cost so far
                    new_h = environment.get_heuristic(neighbor)  # Estimated to goal
                    new_f = new_g + new_h                        # Total estimate
                    print(f"    → {neighbor}: g={new_g}, "
                          f"h={new_h}, f={new_f}")
                    heapq.heappush(pq, (new_f, new_g,
                                        path + [neighbor]))

        return None, float('inf')

    def run_agent(self, environment):
        print("=== A* Search ===")
        result, cost = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result} | Total Cost: {cost}")
        else:
            print("✗ Goal not found.")


# --- Run ---
env = Environment(graph, heuristic, 'A', 'F')
agent = AStarAgent()
agent.run_agent(env)
````

````python
# BEAM Search

import heapq

graph = {
    'A': [('B', 6), ('C', 2)],
    'B': [('D', 7), ('E', 4)],
    'C': [('F', 1), ('G', 5)],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Heuristic: estimated distance to goal F
heuristic = {
    'A': 7,
    'B': 5,
    'C': 2,
    'D': 9,
    'E': 6,
    'F': 0,
    'G': 4
}

class Environment:
    def __init__(self, graph, heuristic, start, goal):
        self.graph = graph
        self.heuristic = heuristic
        self.start = start
        self.goal = goal

    def get_neighbors(self, node):
        return self.graph.get(node, [])

    def get_heuristic(self, node):
        return self.heuristic.get(node, float('inf'))

    def is_goal(self, node):
        return node == self.goal


class BeamSearchAgent:
    def __init__(self, beam_width=2):
        self.beam_width = beam_width     # K — how many nodes to keep

    def action(self, environment):
        # Beam = list of current best paths
        beam = [[environment.start]]
        visited = set()
        visited.add(environment.start)

        print(f"  Beam Width = {self.beam_width}")

        while beam:
            print(f"\n  Current Beam: {[p[-1] for p in beam]}")
            all_neighbors = []    # Collect ALL neighbors from beam nodes

            for path in beam:
                node = path[-1]

                if environment.is_goal(node):
                    return path            # Found goal!

                for neighbor, _ in environment.get_neighbors(node):
                    if neighbor not in visited:
                        h_val = environment.get_heuristic(neighbor)
                        all_neighbors.append((h_val, path + [neighbor]))
                        visited.add(neighbor)

            if not all_neighbors:
                break

            # Sort all neighbors by heuristic, keep only BEST K
            all_neighbors.sort(key=lambda x: x[0])
            print(f"  All candidates: "
                  f"{[(p[-1], h) for h, p in all_neighbors]}")

            beam = [path for _, path in all_neighbors[:self.beam_width]]
            print(f"  Kept in beam (top {self.beam_width}): "
                  f"{[p[-1] for p in beam]}")

            # Check if any beam node is the goal
            for path in beam:
                if environment.is_goal(path[-1]):
                    return path

        return None

    def run_agent(self, environment):
        print("=== Beam Search ===")
        result = self.action(environment)
        if result:
            print(f"\n✓ Goal found! Path: {result}")
        else:
            print("\n✗ Goal not found within beam.")


# --- Run ---
env = Environment(graph, heuristic, 'A', 'F')
agent = BeamSearchAgent(beam_width=2)
agent.run_agent(env)
````

````python
# Hill climbing

import random

# States are numbers, goal is to MAXIMIZE value
# Think of it as finding the highest point on a landscape

class Environment:
    def __init__(self):
        # Each state maps to a value (like altitude on a hill)
        self.state_values = {
            0: 1,
            1: 3,
            2: 7,
            3: 5,
            4: 9,    # Global maximum (our goal peak)
            5: 6,
            6: 2,
            7: 8,
            8: 4,
            9: 0
        }
        self.start_state = 0

    def get_value(self, state):
        return self.state_values.get(state, -1)

    def get_neighbors(self, state):
        # Neighbors are adjacent states
        neighbors = []
        if state - 1 >= 0:
            neighbors.append(state - 1)
        if state + 1 <= 9:
            neighbors.append(state + 1)
        return neighbors

    def is_goal(self, state):
        return self.get_value(state) == max(self.state_values.values())


class HillClimbingAgent:
    def action(self, environment):
        current = environment.start_state
        current_value = environment.get_value(current)

        print(f"  Start State: {current} | Value: {current_value}")

        while True:
            neighbors = environment.get_neighbors(current)
            neighbor_values = [
                (environment.get_value(n), n) for n in neighbors
            ]
            best_value, best_neighbor = max(neighbor_values)

            print(f"  Current: {current}(val={current_value}) | "
                  f"Neighbors: {[(n,v) for v,n in neighbor_values]} | "
                  f"Best Neighbor: {best_neighbor}(val={best_value})")

            # If no neighbor is better, we've reached a peak
            if best_value <= current_value:
                print(f"\n  ▲ Local/Global peak reached at state {current}")
                return current, current_value

            # Move to best neighbor
            current = best_neighbor
            current_value = best_value

    def run_agent(self, environment):
        print("=== Hill Climbing ===")
        result_state, result_value = self.action(environment)
        is_global = environment.is_goal(result_state)
        print(f"\n{'✓ Global' if is_global else '⚠ Local'} "
              f"maximum found!")
        print(f"  State: {result_state} | Value: {result_value}")


# --- Run ---
env = Environment()
agent = HillClimbingAgent()
agent.run_agent(env)
````

````python
import random

# PROBLEM: Find a binary string of length 8 that maximizes
# the number of 1s (target = "11111111")

TARGET = "11111111"
POP_SIZE = 6
GENE_LENGTH = 8
MUTATION_RATE = 0.1
GENERATIONS = 20

class Environment:
    def __init__(self, target):
        self.target = target

    def fitness(self, individual):
        # Count how many positions match the target
        return sum(1 for a, b in zip(individual, self.target) if a == b)

    def is_goal(self, individual):
        return individual == self.target


class GeneticAlgorithmAgent:
    def __init__(self, pop_size, gene_length, mutation_rate):
        self.pop_size = pop_size
        self.gene_length = gene_length
        self.mutation_rate = mutation_rate

    def generate_population(self):
        # Random binary strings as initial population
        population = []
        for _ in range(self.pop_size):
            individual = ''.join(random.choice('01')
                                 for _ in range(self.gene_length))
            population.append(individual)
        return population

    def selection(self, population, environment):
        # Pick top 50% by fitness (elitism selection)
        scored = [(environment.fitness(ind), ind) for ind in population]
        scored.sort(reverse=True)
        print(f"    Fitness scores: {[(s, ind) for s, ind in scored]}")
        # Return top half
        return [ind for _, ind in scored[:self.pop_size // 2]]

    def crossover(self, parent1, parent2):
        # Single-point crossover: split at midpoint
        point = self.gene_length // 2
        child1 = parent1[:point] + parent2[point:]
        child2 = parent2[:point] + parent1[point:]
        return child1, child2

    def mutate(self, individual):
        # Randomly flip bits based on mutation rate
        result = ""
        for gene in individual:
            if random.random() < self.mutation_rate:
                result += '1' if gene == '0' else '0'  # Flip bit
            else:
                result += gene
        return result

    def action(self, environment, generations):
        population = self.generate_population()
        print(f"  Initial Population: {population}")

        for gen in range(generations):
            print(f"\n  --- Generation {gen + 1} ---")

            # Check if solution exists in population
            for individual in population:
                if environment.is_goal(individual):
                    return individual, gen + 1

            # Step 1: Selection — keep the best
            selected = self.selection(population, environment)
            print(f"    Selected (top 50%): {selected}")

            # Step 2: Crossover — breed new children
            children = []
            for i in range(0, len(selected) - 1, 2):
                c1, c2 = self.crossover(selected[i], selected[i + 1])
                children.extend([c1, c2])
                print(f"    Crossover: {selected[i]} + "
                      f"{selected[i+1]} → {c1}, {c2}")

            # Step 3: Mutation — random tweaks
            mutated = [self.mutate(ind) for ind in children]
            print(f"    After Mutation: {mutated}")

            # New population = selected parents + mutated children
            population = selected + mutated

            # Fill up if population too small
            while len(population) < self.pop_size:
                population.append(
                    ''.join(random.choice('01')
                             for _ in range(self.gene_length))
                )

        # Return best from final population
        best = max(population,
                   key=lambda ind: environment.fitness(ind))
        return best, generations

    def run_agent(self, environment, generations):
        print("=== Genetic Algorithm ===")
        print(f"  Target: {environment.target}")
        result, gen = self.action(environment, generations)
        fitness = environment.fitness(result)
        print(f"\n✓ Best solution found in generation {gen}:")
        print(f"  Individual : {result}")
        print(f"  Fitness    : {fitness}/{self.gene_length}")
        if environment.is_goal(result):
            print("  ★ Perfect solution — Goal reached!")


# --- Run ---
random.seed(42)     # For reproducibility
env = Environment(TARGET)
agent = GeneticAlgorithmAgent(POP_SIZE, GENE_LENGTH, MUTATION_RATE)
agent.run_agent(env, GENERATIONS)
```

---

### 🧪 Dry Run (Simplified, 4-bit example)

**Target = "1111", Population size = 4**
```
Initial Population:
  ["0101", "1100", "0011", "1010"]

Generation 1:
  Fitness:  0101→2,  1100→2,  0011→2,  1010→2
  Selected (top 2): ["0101", "1100"]

  Crossover at point 2:
    "0101" + "1100" → "0100", "1101"

  Mutation (flip random bits):
    "0100" → "1100"
    "1101" → "1111"  ← lucky mutation!

  New Population: ["0101","1100","1100","1111"]

Generation 2:
  "1111" found in population → GOAL ✓
```

---

### 📖 Explanation — The 4 Key Steps
```
POPULATION → SELECTION → CROSSOVER → MUTATION → NEW POPULATION
   (all)       (best)      (breed)    (tweak)       (repeat)
````
